```python
import pandas as pd

# Carregar o arquivo CSV
file_path = '/content/TradeDataGolden.csv'  # Verifique se o caminho e o nome do arquivo estão corretos
ghana_data = pd.read_csv(file_path, encoding='latin1')

# Lista de colunas a serem excluídas
columns_to_drop = [
    'Unnamed: 47', 'GrossWgt', 'TypeCode', 'FreqCode', 'RefMonth', 'FlowCode',
    'Partner2ISO', 'Partner2Desc', 'ClassificationSearchCode', 'IsOriginalClassification',
    'CustomsCode', 'CustomsDesc', 'MosCode', 'MotCode', 'MotDesc', 'IsAltQtyEstimated', 'GrossWgt',
    'AltQtyUnitAbbr', 'IsGrossWgtEstimated', 'Cifvalue','LegacyEstimationFlag',
]

# Excluir as colunas
data_reduced = ghana_data.drop(columns=columns_to_drop)

# Salvar o novo DataFrame reduzido em um novo arquivo CSV
output_file_path = '/content/TradeDataGolden_Reduced.csv'  # Verifique se o caminho de saída está correto
data_reduced.to_csv(output_file_path, index=False)

# Exibir as primeiras linhas do DataFrame reduzido
print(data_reduced.head())
```

```python
# Calculate total exports and imports
total_exports = ghana_data[ghana_data['FlowDesc'] == 'Export']['PrimaryValue'].sum()
total_imports = ghana_data[ghana_data['FlowDesc'] == 'Import']['PrimaryValue'].sum()

print(f'Total Exports: {total_exports}')
print(f'Total Imports: {total_imports}')


```

```python
# Identify top trading partners
top_partners = ghana_data['PartnerDesc'].value_counts().head(10)
print('Top Trading Partners:')
print(top_partners)
```

```python
# Identify top exported and imported products
top_exported_products = ghana_data[ghana_data['FlowDesc'] == 'Export']['CmdDesc'].value_counts().head(10)
top_imported_products = ghana_data[ghana_data['FlowDesc'] == 'Import']['CmdDesc'].value_counts().head(10)

print('Top Exported Products:')
print(top_exported_products)
print('Top Imported Products:')
print(top_imported_products)
```

```python


# Definir os países da União Europeia, BRICS e UK
eu_countries = [
    'Austria', 'Belgium', 'Bulgaria', 'Croatia', 'Cyprus', 'Czech Republic', 'Denmark', 'Estonia', 'Finland', 'France',
    'Germany', 'Greece', 'Hungary', 'Ireland', 'Italy', 'Latvia', 'Lithuania', 'Luxembourg', 'Malta', 'Netherlands',
    'Poland', 'Portugal', 'Romania', 'Slovakia', 'Slovenia', 'Spain', 'Sweden'
]
brics_countries = ['Brazil', 'Russia', 'India', 'China', 'South Africa']
uk_countries = ['United Kingdom']

# Criar a nova coluna com base nas condições
def categorize_country(country):
    if country in eu_countries:
        return 'União Europeia'
    elif country in brics_countries:
        return 'BRICS'
    elif country in uk_countries:
        return 'UK'
    else:
        return 'N/A'

data_reduced['Country_Category'] = data_reduced['PartnerDesc'].apply(categorize_country)

# Salvar o DataFrame modificado em um novo arquivo CSV
output_file_path = '/content/TradeDataGolden_Reduced_with_Country_Category.csv'  # Verifique se o caminho de saída está correto
data_reduced.to_csv(output_file_path, index=False)

# Exibir as primeiras linhas do DataFrame modificado
print(data_reduced.head())
```

```python
# Exibir as primeiras 20 linhas do DataFrame
print(data_reduced.iloc[:20])
```

```python
# Calcular a nova coluna para converter trade value em toneladas
data_reduced['Trade_Value_Tons'] = data_reduced['Fobvalue'] / 56_000_000

# Salvar o DataFrame modificado em um novo arquivo CSV
output_file_path = '/content/TradeDataGolden_Reduced_with_Country_Category_and_Trade_Value_Tons.csv'  # Verifique se o caminho de saída está correto
data_reduced.to_csv(output_file_path, index=False)

# Exibir as primeiras linhas do DataFrame modificado
print(data_reduced.head(20))
```
